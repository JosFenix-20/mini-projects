```python
import pandas as pd
path=r"DATOS.csv"
df=pd.read_csv(path)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

x= df[['weight_kg','engine_power_hp','age_years']]
y=df['fuel_consumption_l_per_100km']

x_train,x_test,y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train,y_train)

print(f"Score en conjunto de prueba: {model.score(x_test,y_test)}")

rf_model = RandomForestRegressor(n_estimators=1000, random_state=42)
rf_model.fit(x_train, y_train)

rf_score = rf_model.score(x_test,y_test)
print(f"Score en conjunto de prueba con Random Forest: {rf_score}")
```

```python
import ipywidgets as widgets
from IPython.display import display
import pandas as pd

weight_widget = widgets.FloatText(description="weight", value=1500)
engine_power_widget = widgets.FloatText(description="Power", value=150)
age_widget = widgets.FloatText(description="Age (years)", value=5)
button = widgets.Button(description="Predecir")
output = widgets.Output()

def Pulsar_boton(b):
    with output:
        output.clear_output()
        
        weight = weight_widget.value
        engine_power = engine_power_widget.value
        age = age_widget.value

        input_data = pd.DataFrame(
            [[weight,engine_power,age]],
            columns=['weight_kg','engine_power_hp','age_years'])

        prediction = rf_model.predict(input_data)[0]

        print(f"La prediccion para el consumo de combustible es: {prediction:.2f} L/100km")

button.on_click(Pulsar_boton)

display(weight_widget,engine_power_widget,age_widget,button,output)
```
